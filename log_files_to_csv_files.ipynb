### Subject Ideal csv files

```python
for subject in range(1, 18):
    file_path = dataset_path + '/subject{0}_ideal.log'.format(subject)
    file = pd.read_csv(file_path,header=None, sep='\t',names=columns)
    for column in columns:
        file[column] = file[column].astype('float')
    file['targets'] = file['targets'].astype('int')
    df = file[file['targets'] != 0]
    df.to_csv(r"./data_csv/subject{0}_ideal.csv".format(subject),header=None,sep='\t', index= False)
```
