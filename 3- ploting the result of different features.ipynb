# Comparison between different features bar plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import numpy as np
```

```python
N = 5
Time_domain_accuracy = [92.14, 92.3, 83.22, 72.93, 96.6]
frequency_domain_accuracy = [72.1, 69.1, 62.3, 26.1, 88.7]
wavelet_domain_accuracy = [66.8, 57.7, 48.6, 15.8, 65.2]

barWidth = 0.25


r1 = np.arange(len(Time_domain_accuracy))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 

plt.bar(r1, Time_domain_accuracy, color='#7f6d5f', width=barWidth, edgecolor='white', label='Time domain')
plt.bar(r2, frequency_domain_accuracy, color='#557f2d', width=barWidth, edgecolor='white', label='Frequency domain')
plt.bar(r3, wavelet_domain_accuracy, color='#2d7f5e', width=barWidth, edgecolor='white', label='Wavelet')
 

plt.xlabel('Accuracy of Sensor resutls on different features ', fontweight='bold')


plt.xticks([r + barWidth for r in range(len(Time_domain_accuracy))], ['ACC', 'GYR', 'MAG', 'QUAD', 'ALL'])


plt.legend()
plt.show()
```
